## vp4jl: Visual Programming for JupyterLab
**vp4jl** offers an interactive node-based visual programming (vp) environment. It allows users to build and execute visual workflows without the need for line-by-line textual coding. Seamlessly integrated within the JupyterLab interface, **vp4jl** enables you to enhance your JupyterLab environment with vp capabilities, all without modifying your existing setup.

```python
from torchvision.io import read_image, ImageReadMode
import matplotlib.pyplot as plt
import kornia as K
import numpy as np
# torch.tensor, RGB, channel first, 0-255, shape (C, H, W)
image = read_image('example.jpg', ImageReadMode.RGB)
# To numpy.ndarray, RGB, channel last, 0-255, shape (H, W, C)
plt.imshow(np.transpose(image.numpy(), (1, 2, 0)), cmap='viridis')
plt.axis('off')
plt.show()
# To torch.tensor, RGB, channel first, 0-1, shape (1, 1, H, W)
edge = K.filters.canny(image.unsqueeze(0) / 255)[0]
# To numpy.ndarray, Grayscale, no channel, 0-255, shape (H, W)
plt.imshow(edge.squeeze().numpy(), cmap='gray')
plt.axis('off')
plt.show()
```

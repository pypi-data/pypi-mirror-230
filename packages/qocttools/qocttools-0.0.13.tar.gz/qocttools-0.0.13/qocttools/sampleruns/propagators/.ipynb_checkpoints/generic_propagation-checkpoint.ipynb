```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cmath as cm
import csv
```

```python
from qutip import *
```

```python
#Hamiltonian operators definition from file

H0 = Qobj(np.loadtxt("input/H0.txt", comments = "#", delimiter = ","))
print(H0)

V = Qobj(np.loadtxt("input/V.txt", comments = "#", delimiter = ","))
print(V)
```

```python
#Temporal dependance definition

#from input.f import f
 
dim = H0.shape[0]

w = []
for i in range(dim - 1):
    w.append(H0.data[i+1,i+1] - H0.data[i,i]) 
    
#args is a dictionary that save the parameters of f
args = {"lambda":0.025,  "w1": abs(w[0]), "w2": abs(w[1]), "w3": abs(w[2])}

t_pi_medios = np.pi/(2*args["lambda"]) #pi-pulse duration

#form of f(t)
def f(t, args):
    Lambda = args["lambda"]
    w1 = args["w1"]
    w2 = args["w2"]
    w3 = args["w3"]

    if(t < t_pi_medios):
        return Lambda * np.sin(w1*t)
    elif(t > t_pi_medios and t < 3*t_pi_medios):
        return Lambda * np.sin(w2*t)
    elif(t > 3*t_pi_medios):
        return Lambda * np.sin(w3*t)

#H = H0 + f(t)*V
H = [H0, [V, f]]
```

```python
#simulation run

max_time = 5*t_pi_medios #simulation max time
time_n = 1000 #time discretization (dt)
time = np.linspace(0.0, max_time, time_n) #time steps array
psi0 = basis(dim, 0) #initial state

result = sesolve(H, psi0, time, [], args)
#result = mcsolve(H, psi0, time, [], [], 100, args)
#result = mesolve(H, psi0, time, [], [], args)
```

```python
x_aux = []
x = []
for j in range(dim):
    x_aux.clear()
    for i in range(time_n):
        x_aux.append(abs(result.states[i].data[j, 0])**2)
    x.append(list(x_aux))
    
graph = plt.figure()    
plt.xlabel("time")
plt.ylabel("ocupation")
plt.title("|0> to 0.707*(|0> + |3>) transition based on pi-pulses")
plt.xlim(0, max_time)
plt.ylim(0,1)
plt.grid("best")

for i in range(4):
    plt.plot(time, x[i], label = "|{}>".format(i))

plt.legend()

graph.savefig("output/state_occup.pdf", bbox_inches='tight')
```

```python
#output occup data
file = open("output/data_states_occup.txt", "w")
for i in range(time_n):
    file.write(str(time[i]) + "\t")
    for j in range(dim):    
        file.write(str(x[j][i]) + "\t")
    else:
        file.write("\n")

file.close()
```
